```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Department', hue='Attrition', palette='Set2')
plt.title('Attrition by Department')
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='JobRole', hue='Attrition', palette='coolwarm')
plt.title('Attrition by Job Role')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette='pastel')
plt.title('Monthly Income Distribution by Attrition')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.violinplot(data=df, x='Attrition', y='Age', palette='muted')
plt.title('Age Distribution by Attrition')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='Attrition', y='YearsAtCompany', palette='cool')
plt.title('Years at Company vs Attrition')
plt.show()
```

```python
# Get only numeric columns
num_df = df.select_dtypes(include='number')

# Correlation matrix
plt.figure(figsize=(14,10))
sns.heatmap(num_df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
